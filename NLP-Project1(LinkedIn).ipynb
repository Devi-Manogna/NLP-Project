```python
pd.set_option('display.max_colwidth',100)
data = pd.read_csv(r"C:\Users\Devi_Manogna\Downloads\SMSSpamCollection.csv")
data.columns = ['label','body_text']
data
```

REMOVE PUNCTUATION

```python
# we are practicng to remove the punctuautions frm the above data
import string
string.punctuation
```

```python
#We have to create list comprehension function uses to check through the elemets and returns the result
def remove_punt(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct
data['body_text_clean'] = data['body_text'].apply(lambda x: remove_punt(x))
data.head()
```

TOKENIZATION

```python
#converitng some strings into list of words is called tokenization, here re is regular expressions
import re
def tokenize(text):
    tokens = re.split('\W+',text)
    return tokens
data['tokenized text'] = data['body_text_clean'].apply(lambda x: tokenize(x.lower()))
data.head()
```

REMOVE STOPWORDS

```python
#we have to limit the number of tokens by removing the common words which don't have much meaning like the, is, was etc
import nltk
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenised_list):
    text = [word for word in tokenised_list if word not in stopword]
    return text
data['stopword_removedlist'] = data['tokenized text'].apply(lambda x: remove_stopwords(x))
data.head()
```

STEMMING

```python
ps = nltk.PorterStemmer()
#dir(ps)
```

```python
#this is just one way of representing all the above 3 def functions so here stopword_removedlist = body_text_nostop :)
def clean_text(text):
    text = "".join([word for word in text if word not in string.punctuation])
    tokens = re.split('\W+',text)
    text = [word for word in tokens if word not in stopword]
    return text
data['body_text_nostop'] = data['body_text'].apply(lambda x:clean_text(x.lower()))
data.head()
```

```python
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text
data['body_text_stemmed'] = data['body_text_nostop'].apply(lambda x: stemming(x))
data.head()
```

```python
#stemming just removes the last few words whereas lemma understands the proper meaning before chopping off the word based on real world context meaning
import nltk
wn = nltk.WordNetLemmatizer()
ps = nltk.PorterStemmer()
```

DIFFERENCE OF LEMMATIZER and STEMMING

```python
print(wn.lemmatize("meanness"))
print(wn.lemmatize("meaning"))
```

```python
print(ps.stem("meaning"))
print(ps.stem("meanness"))
```

```python
def lemmatizing(tokenized_text):
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text
data['body_text_lemmatized'] = data['body_text_stemmed'].apply(lambda x: lemmatizing(x))
data.head()
```

VECTORIZATION

```python
#we wont use lambda function here just stemming we did (can do lammetizer but takes time)
def clean_text(text):
    text = "".join([word for word in text if word not in string.punctuation])
    tokens = re.split('\W+',text)
    text = [ps.stem(word) for word in tokens if word not in stopword]
    return text
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(analyzer=clean_text)
x_counts = count_vect.fit_transform(data['body_text'])
print(x_counts.shape)
print(count_vect.get_feature_names_out())
```

```python
#just added space to create a sentence which joining in 3rd step in function and used lambda
def clean_text(text):
    text = "".join([word for word in text if word not in string.punctuation])
    tokens = re.split('\W+',text)
    text = " ".join([ps.stem(word) for word in tokens if word not in stopword])
    return text
data['cleaned_text'] = data['body_text'].apply(lambda x: clean_text(x))
data.head()
```

CountVectorizer - NGrams

```python
#n grams are nothing but just added 2 or 3 or n number of words together below I gave range of 2,2 so it collected all 2 adjacent words together
#the number of columns will just increase as we keep increasing the ngrams values
from sklearn.feature_extraction.text import CountVectorizer
ngram_vect = CountVectorizer(ngram_range=(2,2))
x_counts=ngram_vect.fit_transform(data['cleaned_text'])
print(x_counts.shape)
print(ngram_vect.get_feature_names_out())
```

```python
#trying for small sample data
data_sample = data[0:20]
ngram_vect_sample = CountVectorizer(ngram_range=(2,2))
x_counts_sample = ngram_vect_sample.fit_transform(data_sample['cleaned_text'])
print(x_counts_sample.shape)
print(ngram_vect_sample.get_feature_names_out())
```

```python
x_counts_df = pd.DataFrame(x_counts_sample.toarray())
x_counts_df.columns = ngram_vect_sample.get_feature_names_out()
x_counts_df
```
